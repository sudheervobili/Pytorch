```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# Load pre-trained model (GPT2) and tokenizer
model_name = "gpt2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# Input prompt
prompt = "Once upon a time in a small village,"
input_ids = tokenizer.encode(prompt, return_tensors='pt')

# Generate using Greedy Search
greedy_output = model.generate(input_ids, max_new_tokens=60)
print("\n Greedy Search:")
print(tokenizer.decode(greedy_output[0], skip_special_tokens=True))

# Generate using Beam Search
beam_output = model.generate(input_ids, max_new_tokens=60, num_beams=5, early_stopping=True)
print("\n Beam Search:")
print(tokenizer.decode(beam_output[0], skip_special_tokens=True))

# Generate using Top-p (nucleus) sampling
top_p_output = model.generate(input_ids, max_new_tokens=60, do_sample=True, top_p=0.9, temperature=0.9)
print("\n Top-p Sampling:")
print(tokenizer.decode(top_p_output[0], skip_special_tokens=True))

# Generate using Top-k sampling
top_k_output = model.generate(input_ids, max_new_tokens=60, do_sample=True, top_k=50, temperature=1.0)
print("\n Top-k Sampling:")
print(tokenizer.decode(top_k_output[0], skip_special_tokens=True))
```
